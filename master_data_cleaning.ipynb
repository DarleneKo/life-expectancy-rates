```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = "../life-expectancy-rates/cleaned_data_output.csv"
df = pd.read_csv(file)
```

```python
df1 = df.drop(columns='Unnamed: 0').rename(columns = {"Diphtheria ": "Diphtheria"})
df1
```

### Creating boxplots and histograms to analyze potential outliers:

```python
#Boxplots for diseases
hiv = df1['HIV/AIDS']
diphtheria = df1['Diphtheria']
hepatitisB = df1['Hepatitis B']
measles = df1['Measles']
polio = df1['Polio']

fig1, ax1 = plt.subplots()
ax1.boxplot([hiv, diphtheria, hepatitisB, polio])
plt.xticks([1, 2, 3, 4], ['HIV/AIDS', 'Diphtheria', 'Hepatitis B', 'Polio'])
plt.show()
```

```python
#Measles is separate because it highly skews the image of the boxplots above, check the y-axes
fig1, ax1 = plt.subplots()
ax1.boxplot(measles)
plt.xticks([1], 'Measles')
plt.show()
```

```python
#Boxplots for non-disease related factors
alcohol = df1['Alcohol']
bmi = df1['BMI']
percentage_expenditure = df1['Percentage Expenditure']
schooling = df1['Schooling']

fig1, ax1 = plt.subplots()
ax1.boxplot([alcohol, bmi, schooling])
plt.xticks([1, 2, 3], ['Alcohol', 'BMI', 'Schooling'])
plt.show()
```

```python
fig1, ax1 = plt.subplots()
ax1.boxplot(percentage_expenditure)
plt.xticks([1], 'Percentage Expenditure')
plt.show()
```

```python
#Boxplots for Life expectancies/death rates
life_expectancy = df1['Life Expectancy']
adult_mortality = df1['Adult Mortality']
under_five_deaths = df1['Under-Five Deaths']
infant_deaths = df1['Infant Deaths']

fig1, ax1 = plt.subplots()
ax1.boxplot([life_expectancy, adult_mortality, under_five_deaths, infant_deaths])
plt.xticks([1, 2, 3, 4], ['Life Expectancy', 'Adult Mortality', 'Under-Five Deaths', 'Infant Deaths'])
plt.show()
```
