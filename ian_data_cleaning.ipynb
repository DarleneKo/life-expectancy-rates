```python
import pandas as pd
import matplotlib.pyplot as plt

file = "../cleaned_data_output.csv"
df = pd.read_csv(file)
```

```python
df1 = df.drop(columns='Unnamed: 0').rename(columns = {"Diphtheria ": "Diphtheria"})
df1
```

### Creating boxplots and histograms to analyze potential outliers:

```python
#Set the index as the 'status' of each country
status_df = df1.set_index("Status")

#Separate developed/developing countries
developing = status_df.loc['Developing']
developed = status_df.loc['Developed']

#gather data for each disease for developed and developing countries separately
developed_diph = developed['Diphtheria']
developed_hepB = developed['Hepatitis B']
developed_meas = developed['Measles']
developed_pol = developed['Polio']

developing_diph = developing['Diphtheria']
developing_hepB = developing['Hepatitis B']
developing_meas = developing['Measles']
developing_pol = developing['Polio']

status_df
```

```python
#Create relevant boxplots for developed countries
fig1, ax1 = plt.subplots()
ax1.boxplot([developed_diph, developed_hepB, developed_pol])
plt.xticks([1, 2, 3], ['Diphtheria', 'Hepatitis B', 'Polio'])
plt.show()
```

```python
#Measles is separate due to the high amount of cases, check y-axes
fig1, ax1 = plt.subplots()
ax1.boxplot(developed_meas)
plt.xticks([1], 'Measles')
plt.show()
```

```python
#Creating boxplots for developing countries
fig1, ax1 = plt.subplots()
ax1.boxplot([developing_diph, developing_hepB, developing_pol])
plt.xticks([1, 2, 3], ['Diphtheria', 'Hepatitis B', 'Polio'])
plt.show()
```

```python
#Measles is separate due to the high amount of cases, check y-axes
fig1, ax1 = plt.subplots()
ax1.boxplot(developing_meas)
plt.xticks([1], 'Measles')
plt.show()
```

```python
#Gather data for lifestyle factors
developed_alcohol = developed['Alcohol']
developed_bmi = developed['BMI']

developing_alcohol = developing['Alcohol']
developing_bmi = developing['BMI']

#Developed boxplots
fig1, ax1 = plt.subplots()
ax1.boxplot([developed_alcohol, developed_bmi])
plt.xticks([1, 2], ['Alcohol', 'BMI'])
plt.show()
```

```python
#Developing boxplots
fig1, ax1 = plt.subplots()
ax1.boxplot([developing_alcohol, developing_bmi])
plt.xticks([1, 2], ['Alcohol', 'BMI'])
plt.show()
```

```python
#Gather data for country expenditures
developed_percentage = developed['Percentage Expenditure']
developed_total = developed['Total Expenditure']

developing_percentage = developing['Percentage Expenditure']
developing_total = developing['Total Expenditure']

#Generate boxplots for Developed countries' expenditures
fig1, ax1 = plt.subplots()
ax1.boxplot([developed_percentage, developed_total])
plt.xticks([1, 2], ['Percentage Expenditure', 'Total Expenditure'])
plt.show()
```

```python
#Generate boxplots for developing countries' expenditures
fig1, ax1 = plt.subplots()
ax1.boxplot([developing_percentage, developing_total])
plt.xticks([1, 2], ['Percentage Expenditure', 'Total Expenditure'])
plt.show()
```

```python
#Gather data for adult/infant mortality rates
developed_adult = developed['Adult Mortality']
developed_infant = developed['Infant Deaths']

developing_adult = developing['Adult Mortality']
developing_infant = developing['Infant Deaths']
```

```python
#Create boxplots for developed countries' mortality rates
fig1, ax1 = plt.subplots()
ax1.boxplot([developed_adult, developed_infant])
plt.xticks([1, 2], ['Adult Mortality', 'Infant Deaths'])
plt.show()
```

```python
#Generate boxplots for developing countries' mortality rates
fig1, ax1 = plt.subplots()
ax1.boxplot([developing_adult, developing_infant])
plt.xticks([1, 2], ['Adult Mortality', 'Infant Deaths'])
plt.show()
```

```python
#Gather average life expectancies from 2000 - 2015 for developed/developing countries
time_group_dev = developed.groupby('Year')['Life Expectancy'].mean()
time_group_deving = developing.groupby('Year')['Life Expectancy'].mean()
```

```python
#Create boxplots for average life expectancies for developed/developing countries
fig1, ax1 = plt.subplots()
ax1.boxplot([time_group_dev, time_group_deving])
plt.xticks([1, 2], ['Developed', 'Developing'])
plt.ylabel("Average Life Expectancy")
plt.show()
```
